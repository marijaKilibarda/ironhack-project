```python
def check_df(data, head=5):
    print(f'Dimensions of this dataset : {data.shape}')
    print("**********************************************************")
    
    # Display first few rows
    print(f"First {head} row of the dataset:\n")
    print(data.head(head).to_string(index=False))  # Display the data rows formatted
    print("\n**********************************************************")
    print(f"Numerical information about this dataset:\n{data.describe()}")
    print("\n**********************************************************")
    
    print(f"General information about the dataset : {data.info()}")
    print("\n**********************************************************")
    
    print(f"How many total null values ​​are there in this data set?\n{data.isnull().sum()}")
    print("\n**********************************************************")
    
    print(f"Are there any repeated observations in this data set?\n{data.duplicated().sum()}")
    print("\n**********************************************************")
    
    print(f"How many total unique values are there in this data set?\n{data.nunique()}")
check_df(df_01)
check_df(df_02)
```

```python
plt.figure(figsize=(6, 6))
df_01['Operating System'].value_counts().plot.pie(autopct='%1.1f%%', 
                                                  startangle=90, 
                                                  colors=['firebrick', 'steelblue'],
                                                  legend=True) 

plt.title('Operating System Distribution')
plt.ylabel('')
plt.legend(title='Operating Systems') 
plt.show()
```

```python
# Android commands a striking 79.1% share, clearly establishing it as the dominant choice among users. 
# This vast majority underscores Android’s widespread appeal and reach. 
# Meanwhile, iOS captures a solid 20.9%, demonstrating that although fewer, 
# iOS users still represent a significant and loyal audience in the mobile ecosystem.
```

```python
plt.figure(figsize=(8, 4))
sns.countplot(data=df_02, 
              x='Gender',
              hue='OS',
              palette=['firebrick', 'steelblue', 'lightgrey', 'darkgrey'] )

plt.title('Operating System Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='OS')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=df_01, 
              x='Device Model',
              hue='Gender',
              palette=['lightcoral','powderblue']
             )

plt.title('Device Model Distribution by Gender')
plt.xlabel('Device Model')
plt.ylabel('Count')
plt.legend(title='GENDER')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# The graph paints a vivid picture of device preferences across genders, spotlighting clear trends in model choice. 
# For women, the Xiaomi Mi11 and iPhone 12 shine as the top picks, marking them as the standout favorites. On the other hand, 
# a broader range of phone brands sees higher popularity among men, underscoring distinct differences in device loyalty and brand preference by gender.
```

```python
plt.figure(figsize=(20, 6))
sns.boxplot(data=df_01,
            x='Age', 
            y='App Usage Time (min/day)', 
            hue='Gender',  
            palette=['lightcoral','powderblue']
            )
plt.title('App Usage Time by Age and Gender')
plt.show()
```

```python
#heavy App usage time of a male groupe 35+ to 56/57 years compared to female
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df_01, 
                x='Age', 
                y='Data Usage (MB/day)', 
                hue='Operating System', 
                palette=['r', 'blue'], 
                alpha=0.6)
plt.title('Age vs. Data Usage')
plt.xlabel('Age')
plt.ylabel('Data Usage (MB/day)')
plt.show()
```

```python
# The user engagement and behavior patterns are more influenced by personal habits 
# or app preferences than age demographics.
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df_01, 
                x='Battery Drain (mAh/day)', 
                y='Data Usage (MB/day)', 
                hue='Operating System', 
                palette=['r', 'blue'], 
                alpha=0.6)
plt.title('Battery Drain vs. Data Usage')
plt.xlabel('Battery Drain')
plt.ylabel('Data Usage (MB/day)')
plt.show()
```

```python
# Data show slightly higher consumption among iOS users, 
# potentially reflecting higher engagement with resource-intensive apps
```
